```python
#Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd

# Load the dataset 'worldbank.xlsx' from the raw GitHub link
df=pd.read_excel('https://raw.githubusercontent.com/Knowasiak/A2_World_Region_Analysis/main/worldbank.xlsx')
df.head()
df.info()
```

```python
# Convert 'Technology Exports' to million USD
df['Technology Exports'] = df['Technology Exports'] / 1_000_000

# Convert 'Population' to million
df['Population'] = df['Population'] / 1_000_000

# Convert 'Working Population' to million
df['Working Population'] = df['Working Population'] / 1_000_000

print("DataFrame after converting units to millions:")
display(df.head())
df.info()
```

```python
# Extracting only the South American and Caribbean countries from the full dataset ensuring the dataset aligns with the assigned region for analysis.

region_df = df[df["Region"].isin(["Southern Americas and Caribbean"])]

#Preview of the first 5 rows to check the filtered dataset.
#Review the summary statistics of the dataset.
region_df.head()
region_df.info()
region_df.describe()
```

```python
 region_mask = df['Region'] == 'Southern Americas and Caribbean'
 reg = df[region_mask].copy()
```

```python
 num_cols = df.select_dtypes(include=[np.number]).columns
 world_mean = df[num_cols].mean()
 reg_mean = reg[num_cols].mean()
```

```python
 missing_counts = reg.isna().sum().sort_values(ascending=False)
 reg_num = reg[num_cols]
 reg_means = reg_num.mean()
 distances = ((reg_num - reg_means)**2).sum(axis=1)**0.5
 reg['distance_to_mean'] = distances
```

```python
# Identify numeric columns
num_cols = df.select_dtypes(include=[np.number]).columns

# Compute world and regional means
world_mean = df[num_cols].mean()
reg_mean = reg[num_cols].mean()

comp = pd.DataFrame({
    "world_mean": world_mean,
    "region_mean": reg_mean
})
comp["diff"] = comp["region_mean"] - comp["world_mean"]

# Pick a few interesting metrics for plotting
metrics_to_plot = [
    "GDP PPP",
    "Youth Literacy Rate",
    "Population (Urban %)",
    "Technology Exports"
]

plot_df = comp.loc[metrics_to_plot]

# Bar chart: region vs world
plot_df[["world_mean", "region_mean"]].plot(
    kind="bar",
    figsize=(8, 5)
)
plt.title("Southern Americas & Caribbean vs World Averages")
plt.ylabel("Average Value")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
